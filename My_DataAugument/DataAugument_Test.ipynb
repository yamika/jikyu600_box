```python
# -*- coding: utf-8 -*-
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from My_DataAugument import *
```

```python
img = Image.open("./limitspeed-00007.jpg").convert('RGB')
plt.imshow(np.array(img))
```

```python
# 画像の反転
out_img = flip_image(img,flip_type='LR')
plt.imshow(np.array(out_img))
```

```python
out_img = flip_image(img,flip_type='TB')
plt.imshow(np.array(out_img))
```

```python
# 画像の回転
out_img = rotate_image(img,rotate_value=30)
plt.imshow(np.array(out_img))
```

```python
# 画像を右にずらす
out_img = move_image(img,move_value=10,move_type='right')
plt.imshow(np.array(out_img))
```

```python
# 画像を上にずらす
out_img = move_image(img,move_value=30,move_type='top')
plt.imshow(np.array(out_img))
```

```python
# 画像のコントラストを上げる
out_img = change_contrast(img,contrast_value=1.2)
plt.imshow(np.array(out_img))
```
